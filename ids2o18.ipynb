```python
def encode_labels(df):
  df = pd.get_dummies(data = df, columns=["Label"], dtype=np.uint8)
  return df
```

```python
def drop_infinate_null(df):
    print (df.shape)

    # replace infinity value as null value
    print("Count of NaN values              : " + str(df.isna().sum().sum()))
    df = df.replace(["Infinity", "infinity"], np.inf)
    df = df.replace([np.inf, -np.inf], np.nan)
    print("Count of Infinity and NaN values : " + str(df.isna().sum().sum()))

    # drop all null values
    df.dropna(inplace=True)
    doubBef = df.shape[0]
    df.drop_duplicates(inplace=True)
    print("Count of duplicates              : " + str(doubBef - df.shape[0]))

    return df
```

```python
def reduce_memory_usage(prop):
  start_mem_usg = prop.memory_usage().sum() / 1024**2
  print("Memory usage of properties dataframe is :",start_mem_usg," MB")
  for col in prop.columns:
    if prop[col].dtype == object:
      continue
    mx = prop[col].max()
    mn = prop[col].min()

    asint = prop[col].fillna(0).astype(np.int64)
    result = (prop[col] - asint)
    result = result.sum()
    if result > -0.01 and result < 0.01:
        prop[col] = prop[col].astype(int)

    if prop[col].dtype == int:
      if mn >= 0:
        if mx < np.iinfo(np.uint8).max:
          prop[col] = prop[col].astype(np.uint8)
        elif mx < np.iinfo(np.uint16).max:
          prop[col] = prop[col].astype(np.uint16)
        elif mx < np.iinfo(np.uint32).max:
          prop[col] = prop[col].astype(np.uint32)
        else:
          prop[col] = prop[col].astype(np.uint64)
      else:
        if mn > np.iinfo(np.int8).min and mx < np.iinfo(np.int8).max:
          prop[col] = prop[col].astype(np.int8)
        elif mn > np.iinfo(np.int16).min and mx < np.iinfo(np.int16).max:
          prop[col] = prop[col].astype(np.int16)
        elif mn > np.iinfo(np.int32).min and mx < np.iinfo(np.int32).max:
          prop[col] = prop[col].astype(np.int32)

    elif prop[col].dtype == float:
      prop[col] = prop[col].astype(np.float32)


  print("___MEMORY USAGE AFTER COMPLETION:___")
  mem_usg = prop.memory_usage().sum() / 1024**2
  print("Memory usage is: ",mem_usg," MB")
  print("This is ",100*mem_usg/start_mem_usg,"% of the initial size")
  return prop
```

```python
def reduce_dimensions_with_corr(df):
  df_corr = df.corr()
  while df_corr.isna().all().any():
    df_corr = df.corr()
    df.drop(columns=df_corr.columns[df_corr.isna().all()], inplace=True)
    print(df_corr.columns[df_corr.isna().all()])
  
  del df_corr
```

```python
def do_preproccessing(df):
  res = []
  for d in df:
    remove_unneccecary_colm(d)
    d = define_object_types(d)
    d = drop_infinate_null(d)
    d = reduce_memory_usage(d)
    res.append(d)
  return res
```
